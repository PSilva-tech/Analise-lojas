### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
soma1 = loja['Preço'].sum()
valor1 = f'{soma1:,.2f}'.replace(",", "X").replace(".", ",").replace("X", ".")
print(f'Total vendido pela loja 1: R$ {valor1}')
```

```python
soma2 = loja2['Preço'].sum()
valor2 = f'{soma2:,.2f}'.replace(",", "X").replace(".", ",").replace("X", ".")
print(f'Total vendido pela loja 2: R$ {valor2}')
```

```python
soma3 = loja3['Preço'].sum()
valor3 = f"{soma3:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
print(f"Total vendido pela loja 3: R$ {valor3}")
```

```python
soma4 = loja4['Preço'].sum()
valor4 = f"{soma4:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
print(f"Total vendido pela loja 4: R$ {valor4}")
```

```python
soma_total = soma1 + soma2 + soma3 + soma4

print(f"Soma total dos preços: R$ {soma_total:.2f}")

```

```python
valor = 5871490.79
valor_formatado = f"{valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
print(f"Valor formatado: R$ {valor_formatado}")
```

```python
import matplotlib.pyplot as plt

total_vendido = {
    'Loja 1': soma1,
    'Loja 2': soma2,
    'Loja 3': soma3,
    'Loja 4': soma4
}

plt.figure(figsize=(10, 6))
plt.bar(total_vendido.keys(), total_vendido.values(), color=['skyblue', 'lightgreen', 'salmon', 'lightcoral'])

plt.title('Total Vendido por Loja', fontsize=16)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Total Vendido (R$)', fontsize=12)

plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: f'R$ {x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')))

plt.show()
```

# 2. Vendas por Categoria
